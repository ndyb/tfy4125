## Oppgave 2: Harmonisk oscillator - frie svingninger

Newton's andre lov gir

$$
\begin{align}
F =& ma \\
m\ddot{x} =& -kx \\
m\ddot{x} + kx =& 0
\end{align}
$$

Har oppgitt at en generell løsning er gitt ved

$x=A\cos(\omega t + \theta)$

Det gir 

$\dot{x}=-A\omega \sin(\omega t + \theta)$ 

og 

$\ddot{x}=-A\omega^2 \cos(\omega t + \theta)$



Setter inn initialverdibetingelsene våre og får

$x(0) = A\cos(\theta) \implies A = \frac{x(0)}{\cos(\theta)}$

$\dot{x}(0) = -A\omega \sin(\theta) = -\frac{x(0)}{\cos(\theta)}\omega \sin(\theta) = - x(0) \tan(\theta) \omega$



Som gir

$\theta = \arctan(-\frac{\dot{x}(0)}{\omega x(0)})$

For å løse ligningene må vi finne $\omega$

$$
\begin{align}
m\ddot{x} + kx =& 0 \\
-mA\omega^2 \cos(\omega t + \theta) + kA \cos(\omega t + \theta) =& 0 \\
-m\omega^2+k=&0 \\
\omega =& \sqrt{\frac{k}{m}}
\end{align}
$$

$\theta = \arctan(\frac{-\dot{x}(0) \sqrt{m}}{x(0) \sqrt{k}})$

Oppgitte konstanter

```python
m = 0.01
k = 0.1
x_0 = 0.01
dx_0 = 0.05
```

### Analytisk løsning av diffligningen

```python
theta = arctan(-(dx_0*sqrt(m))/(x_0*sqrt(k)))
print(theta)
```

```python
A = x_0 / cos(theta)
print(A)
```

```python
omega = sqrt(k/m)
print(omega)
```

```python
f = lambda t: A*cos(omega*t+theta)
```

$\omega = \frac{2 \pi}{T} \implies T = \frac{2 \pi}{\omega}$

```python
T = 2 * pi / omega
print(T)
```

### Numerisk løsning av diffligningen

Parametre for løsning

```python
n = 50000
t_min = 0
t_max = 4*T
```

```python
t = linspace(t_min, t_max, n)
h = t[1] - t[0]

x = zeros(n)
x[0] = x_0
u = zeros(n)
u[0] = dx_0

du = lambda x: (-k/m)*x

for i in range(n-1):
    u[i+1] = u[i] + h * du(x[i])
    x[i+1] = x[i] + h * u[i]
```

### Plotting av løsningene

```python
plt.figure()
plt.axis([t_min, t_max, min(f(t))*1.1, max(f(t))*1.1])
plt.plot(t, x, label='Numerisk løsning')
plt.plot(t, f(t), label='Analytisk løsning')
plt.legend(loc = 'lower right')
plt.title('Harmonisk bevegelse uten demping')
print("n: " + str(len(x)))
```

Når tidsintervallet er $\frac{4T}{50000}$ så er den numeriske løsningen relativt lik den analytiske på intervallet

## Oppgave 3: Harmonisk oscillator - dempede svingninger

```python
b = 0.001
```

```python
x = zeros(n)
x[0] = x_0
u = zeros(n)
u[0] = dx_0

for i in range(n-1):
    u[i+1] = u[i] - b*h*u[i]/m - k*h*x[i]/m
    x[i+1] = x[i] + h * u[i]
```

### Analytisk løsning

Analytisk løsning er gitt av

$x=A \mathrm{e}^{\frac{-b}{2m}t} \cos(\omega t + \theta)$

```python
omega_1 = sqrt(omega**2-(b**2/4))
f = lambda t: A*e**((-b*t)/(2*m))*cos(omega_1*t+theta)
```

```python
plt.figure()
plt.plot(t, x, label='Numerisk løsning')
plt.plot(t, f(t), label='Analytisk løsning')
plt.legend(loc = 'lower right')
plt.title('Harmonisk bevegelse med demping')
```

```python
zeta = b / (2*sqrt(m*k))
print(zeta)
```

Når $\zeta < 1$ så er systemet underdempet 

For å få et kritisk dempet system må $\zeta = 1$

```python
b = 2*sqrt(m*k)
```

```python
x = zeros(n)
x[0] = x_0
u = zeros(n)
u[0] = dx_0

for i in range(n-1):
    u[i+1] = u[i] - b*h*u[i]/m - k*h*x[i]/m
    x[i+1] = x[i] + h * u[i]
    
plt.figure()
plt.plot(t, x)
plt.title('Harmonisk bevegelse med kritisk demping')
```

## Oppgave 4: Harmonisk oscillator - drevne svingninger

```python
wd = 7
F_0 = 0.01
```

Vi får følgende diffligning

$m\ddot{x}+b\dot{x}+kx=F_0 \cos(\omega_d t)$

Skal dele den opp i to første-ordens diskrete diffligninger

$u' = \ddot{x}$

$u = \dot{x}$

$u_n = \frac{x_{n+1}-x_n}{h} \implies x_{n+1} = x_n + u_n h$

$\ddot{x}=-\frac{b}{m}\dot{x}-\frac{k}{m}x+F_0 \cos(\omega_d t) = u'$

$\frac{u_{n+1} - u_n}{h}=\ddot{x}_n \implies u_{n+1}=-\frac{bh}{m}\dot{x}_n-\frac{kh}{m}x_n+F_0 \cos(\omega_d n h) + u_n$

```python
x = zeros(n)
x[0] = x_0
u = zeros(n)
u[0] = dx_0

for i in range(n-1):
    u[i+1] = u[i] - b*h*u[i]/m - k*h*x[i]/m + F_0 * cos(wd * i * h)
    x[i+1] = x[i] + h * u[i]
    
plt.figure()
plt.plot(t, x)
plt.title('Harmonisk bevegelse med kritisk demping og ytre kraft')
```

Når vi har en en harmonisk bevegelse med demping så blir leddet som styrer den innebygde harmonisk bevegelsen multiplisert med $\mathrm{e}^{-kt}$. Når $t\to \infty$ så blir dette leddet $0$, og det eneste gjenstående leddet er den den ytre kraften som driver systemet. Hvis dempingen er høy så vil man raskere nå steady-state løsningen som er gitt av den ytre kraften alene. Transient løsningen er hele løsningen på problemet fra $t=0$ til $t=\infty$.

```python
wd = omega_1
plt.figure()
for j in [0.5, 0.9, 1.0, 1.1, 2.0]:        
    x = zeros(n)
    x[0] = x_0
    u = zeros(n)
    u[0] = dx_0
    
    for i in range(n-1):
        u[i+1] = u[i] - b*h*u[i]/m - k*h*x[i]/m  + F_0 * cos(wd * i * h * j)
        x[i+1] = x[i] + h * u[i]
    
    plt.plot(t, x, label=str(round(wd*j, 2)))
    
plt.legend(loc='lower left')
plt.title('Harmonisk bevegelse med 10% av kritisk demping og ytre kraft som virker med forskjellig frekvens')
```

Resonansfrekvensen er gitt av $\omega_1$.
