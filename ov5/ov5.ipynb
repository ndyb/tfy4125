```python
n = 5000
x = linspace(0, 2*pi, n)
f = lambda x: (1+cos(x))/2
plt.figure()
plt.grid()
plt.xticks(np.arange(min(x), max(x), pi/2))
plt.plot(x, f(x))
```

Når $\theta=0$ så kjører begge legemer parallellt, og siden de har samme fart så må de allerede være kollidert. Derfor er forholden mellom kinetisk energi før og etter lik. Når $\theta=\pi$ så kjører legemene inn i hverandre, og siden de har samme masse og fart så stopper de helt opp og den kinetiske energien blir null. Når $\theta=\frac{1}{2} \pi$ eller $\theta=\frac{3}{2} \pi$ så treffer legemene hverandre vinkelrett, og den kinetiske energien blir halvert etter kollisjonen.
